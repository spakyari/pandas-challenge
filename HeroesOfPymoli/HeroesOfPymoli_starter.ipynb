```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# Extract unique players SN.
players = purchase_data['SN'].unique()

# Create a dictionary to store players' count
myDict=[{"Total Players": players.size}]

total_players = pd.DataFrame(myDict)
total_players 


```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data['Item Name'].unique()
items_count = unique_items.size

avarage_price = round(purchase_data['Price'].mean(),2)

purchase_count = purchase_data.index.stop

total_revenue = round(purchase_data['Price'].sum(),2)

purchase_analysis = pd.DataFrame([
    {'Number of Unique Items':items_count,
     'Average Price':avarage_price,
        'Number of Purchases':purchase_count,
     'Total Revenue':total_revenue
    }
])


# Format cells
purchase_analysis['Average Price'] = purchase_analysis['Average Price'].map("${:,.2f}".format)
purchase_analysis['Total Revenue'] = purchase_analysis['Total Revenue'].map("${:,.2f}".format)


purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

# Create a base table with duplicate SN's and respective genders
uniplayers_df = purchase_data.loc[:,['SN', 'Gender']]

# Eliminate duplicate rows
uniplayers_df.drop_duplicates(inplace=True)

# Group SNs by Gender and aggregate count total Sn's under each gender
genders_df = uniplayers_df.groupby('Gender').count()

# Rename 'SN' column header to 'Total Count'
genders_df.rename(columns = {'SN':'Total Count'},inplace= True)

# Add 'Percentage of Players' column and calculate precentages
genders_df['Percentage of Players'] = round((genders_df['Total Count']/ genders_df['Total Count'].sum())*100,2)

# Format the 'Percentage of Players' column
genders_df['Percentage of Players'] = genders_df['Percentage of Players'].map("{:,.2f}%".format)

#Show ouput dataframe
genders_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

PAG_df = purchase_data.loc[:,['Gender','Price']].groupby('Gender').agg(['count','mean', 'sum'])


PAG_df= pd.merge(genders_df,PAG_df,on = 'Gender')

PAG_df.columns

PAG_df['Avg Total Purchase per Person'] = round(PAG_df[('Price', 'sum')] / PAG_df['Total Count'],2)

PAG_df.drop(axis=1, inplace= True, columns = ['Total Count','Percentage of Players'])

PAG_df.rename(inplace = True,columns = {('Price', 'count'): 'Purchase Count', 
                         ('Price', 'mean'): 'Average Purchase Price',
                         ('Price', 'sum'):'Total Purchase Value'
                        })

PAG_df['Average Purchase Price'] = PAG_df['Average Purchase Price'].map("${:,.2f}".format)
PAG_df['Total Purchase Value'] = PAG_df['Total Purchase Value'].map("${:,.2f}".format)
PAG_df['Avg Total Purchase per Person'] = PAG_df['Avg Total Purchase per Person'].map("${:,.2f}".format)


PAG_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create Base table.
AD_df = purchase_data.loc[:,['SN','Age']]

# Drop duplicate rows.
AD_df.drop_duplicates(inplace=True)

# Drop redundant column - SN.
AD_df.drop(columns=['SN'], inplace=True)

# Create age group bins.
bins = [0, 9, 14, 19, 24, 29, 34,39, 100]

# Create age group labels.
mylabels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Add column with respective age groups.
AD_df['Age Group'] = pd.cut(AD_df['Age'], bins=bins, labels = mylabels,include_lowest=True)

# Group by age group and count each age group.
AD_df=AD_df.groupby('Age Group').count()

# Add column with respective age groups percentages.
AD_df['Percentage of Players'] = round(AD_df['Age']/AD_df['Age'].sum()*100,2)

# Polish and format table
AD_df.rename(inplace= True, columns = {'Age': 'Total Count'})

AD_df['Percentage of Players'] = AD_df['Percentage of Players'].map("{:,.2f}%".format)

AD_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

PAA_df = purchase_data.loc[:,['Age','Price']]

PAA_df['Age Group'] = pd.cut(PAA_df['Age'], bins=bins, labels = mylabels,include_lowest=True)

PAA_df.drop(columns= 'Age',inplace=True)

PAA_df = PAA_df.groupby('Age Group').agg(['count','mean', 'sum'])

PAA_df.columns = PAA_df.columns.droplevel(0)

PAA_df= pd.merge(AD_df,PAA_df,on = 'Age Group')


PAA_df['Avg Total Purchase per Person'] = round(PAA_df[('sum')] / PAA_df['Total Count'],2)

PAA_df.drop(axis=1, inplace= True, columns = ['Total Count','Percentage of Players'])



PAA_df.rename(inplace = True,columns = {'count': 'Purchase Count', 
                         'mean': 'Average Purchase Price',
                         'sum':'Total Purchase Value'
                         })

#PAA_df['Avg Total Purchase per Person'] = round(PAA_df[('Total Purchase Value')] / PAA_df['Purchase Count'],2)

PAA_df['Average Purchase Price'] = PAA_df['Average Purchase Price'].map("${:,.2f}".format)
PAA_df['Total Purchase Value'] = PAA_df['Total Purchase Value'].map("${:,.2f}".format)
PAA_df['Avg Total Purchase per Person'] = PAA_df['Avg Total Purchase per Person'].map("${:,.2f}".format)


PAA_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
TS_df = purchase_data.loc[:,['SN','Price']].groupby('SN').agg(['count','mean', 'sum'])


TS_df.columns=TS_df.columns.droplevel(0)

TS_df.sort_values('sum',ascending = False,inplace= True)

TS_df.rename(inplace = True, columns = {'count':'Purchase Count',
                                        'mean':'Average Purchase Price',
                                       'sum':'Total Purchase Value'})

TS_df['Average Purchase Price'] = TS_df['Average Purchase Price'].map("${:,.2f}".format)
TS_df['Total Purchase Value'] = TS_df['Total Purchase Value'].map("${:,.2f}".format)

TS_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
input_df = purchase_data.loc[:,['Item ID','Item Name','Price']]


MPI_df = input_df.groupby([ 'Item ID','Item Name']).agg(['count','min','sum'])

MPI_df.rename(inplace= True, columns = {'count':'Purchase Count' ,'min':'Item Price','sum':'Total Purchase Value'}) 

MPI_df.columns=MPI_df.columns.droplevel(0)

MPS_df = MPI_df.copy()

MPI_df['Item Price'] = MPI_df['Item Price'].map("${:,.2f}".format)
MPI_df['Total Purchase Value'] = MPI_df['Total Purchase Value'].map("${:,.2f}".format)


MPI_df.sort_values('Purchase Count',ascending = False,inplace= True)

MPI_df.head()
```
